En keras usamos la palabra `Dense` para definir el `layer` o capa de neuronas conectadas.

Como podemos ver solo hay una capa conectada
```python
keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```
<br>
Hay dos roles de función que debemos tener en cuenta y estas son funciones de pérdida o `loss functions` y optimizadores o `optimizers`.
<br>

Nos gusta pensar de esta manera:
Las redes neuronales no tienen idea entre la correlación de `X` y `Y` por lo tanto lo tienen que adivinar.

Digamos que adivina que Y es igual a 10X menos 10. Luego usará los datos que conoce, ese es el conjunto de X y Y que ya hemos visto para medir qué tan bueno o qué tan malo fue su suposición. La función de pérdida ó `loss functions` mide esto y luego le da los datos al optimizador, que determina la siguiente suposición.

Entonces la lógica es que cada conjetura debe ser mejor que la anterior.

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

Para `Xs` y `Ys` debemos de crear un vector, para eso usamos la librería numpy la cual nos permite crear arrays de forma sencilla

```python
import numpy as np
xs = np.arange(-1, 5, dtype=float)
ys = np.array([-3, -1, 1, 3, 5, 7], dtype=float)
```
posteriormente debemos de 

````
import tensorflow as tf
import numpy as np
from tensorflow import keras

model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.arange(-1, 5, dtype=float)
ys = np.array([-3, -1, 1, 3, 5, 7], dtype=float)

model.fit(xs, ys, epochs=500)
````

````
print(model.predict([10.0]))
````
